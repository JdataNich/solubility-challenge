# A Model to Predict Drug Solubility 

## Importing Libraries 

```python
import pandas as pd
import sklearn 
from sklearn import preprocessing 
from sklearn import linear_model
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
# import training and test data sets
training_csv = pd.read_csv('C:/Users/joe/Downloads/training_full_set.csv')
```

```python
#dropping null values
training_df_drop = training_csv.dropna(axis=1, how='any')
#setting index to compounds name
training_df = training_df_drop.set_index('NAME')
```

```python
#just using numeric data (dropping smiles and inchi keys)
df = training_df._get_numeric_data()
```

```python
#new dataframe made up of everything but solubility
#and an array for input into feature reduction algorithm 
X_df = df.drop('S0 (mM)', axis=1)
Y_df = df.loc[:,'S0 (mM)']
X = X_df.values
y = Y_df.values
```

```python
#using two different feature reduction methods 
lasso = Lasso()
elastic = ElasticNet()

lasso_selector = RFE(lasso, 14)
elastic_selector = RFE(elastic, 14)

lasso_model_select = lasso_selector.fit(X, y)
elastic_model_select = elastic_selector.fit(X, y)
```

```python
X_new_lasso = lasso_model_select.transform(X)
X_new_elastic = elastic_model_select.transform(X)

print(lasso_model_select.get_support())
print(elastic_model_select.get_support())
```

```python
X_lasso_columns = X_df.columns[lasso_model_select.get_support()]
X_elastic_columns = X_df.columns[elastic_model_select.get_support()]
print(X_lasso_columns)
print(X_elastic_columns)
```
