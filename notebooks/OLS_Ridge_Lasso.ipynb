```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from jupyterthemes import jtplot
```

```python
FILE = r'~\Projects\solubility-challenge\data\training_subset.csv'
FILE_TEST = r'~\Projects\solubility-challenge\data\test_subset.csv'
jtplot.style(theme='gruvboxd')
```

```python
df_train = pd.read_csv(FILE)
df_test = pd.read_csv(FILE_TEST, na_values = 'TOO SOLUBLE TO MEASURE')
df_test = df_test.dropna(axis = 0)
```

```python
df_train.head()
```

```python
df_test['S0_Mm'] = (df_test['S (_g/L)'] / df_test['MW']) * 1000
```

```python
# using Log S as a y vaue to normalize distribution
df_train['Log_S'] = df_train['S0 (mM)'].apply(lambda x : np.log10(x))
df_test['Log_S'] = df_test['S0_Mm'].apply(lambda x : np.log10(x))
```

```python
# nROH, nOHP, nHDon, nHAcc show low variance
# Will feature engineer by summing them all as electronegative/positive groups
df_train['nEnegEpos'] = df_train['nROH'] + df_train['nOHp'] + df_train['nHDon'] + df_train['nHAcc']
df_test['nEnegEpos'] = df_test['nROH'] + df_test['nOHp'] + df_test['nHDon'] + df_test['nHAcc']
```

```python
predictors = ['nEnegEpos','MW','Sv','Si','Hy','MLOGP']
target = 'Log_S'
```

```python
ols_model = LinearRegression()
```

```python
X_train = df_train[predictors]
y_train = df_train[target]

X_test = df_test[predictors]
y_test = df_test[target]
```

```python
ols_model.fit(X_train, y_train)
```

```python
y_predict = ols_model.predict(X_test)
```

```python
df_test['OLS_Predict'] = y_predict
```

```python
ols_mse = mean_squared_error(y_test, y_predict)
ols_r2 = r2_score(y_test, y_predict)
```

```python
print(f'The mean squared error for the scikit learn OLS model is {ols_mse}')
print(f'The R squared for the scikit learn OLS model is {ols_r2}')
```

## Ridge & Lasso models

```python
ss = StandardScaler()
```

```python
results_ridge = {}
reg_path_ridge = {}

for alpha in np.linspace(0,20):
    # Scale the data
    X_train_scaled = ss.fit_transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    #fit the model
    model = Ridge(alpha = alpha)
    model.fit(X_train_scaled, y_train)
    ridge_params = dict(zip(predictors, model.coef_))
    reg_path_ridge[alpha] = ridge_params
    
    # Find the results on the test set
    Ridge_Predict = model.predict(X_test_scaled)
    mse_ridge = mean_squared_error(y_test, Ridge_Predict)
    r2_ridge = r2_score(y_test, Ridge_Predict)
    results_ridge[alpha] = mse_ridge
```

```python
results_lasso = {}
reg_path_lasso = {}

for alpha in np.linspace(0,0.1):
    # Scale the data
    X_train_scaled = ss.fit_transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    #fit the model
    model = Lasso(alpha = alpha)
    model.fit(X_train_scaled, y_train)
    lasso_params = dict(zip(predictors,model.coef_))
    reg_path_lasso[alpha] = lasso_params
    
    # Find the results on the test set
    Lasso_Predict = model.predict(X_test_scaled)
    mse_lasso = mean_squared_error(y_test, Lasso_Predict)
    r2_lasso = r2_score(y_test, Lasso_Predict)
    results_lasso[alpha] = mse_lasso
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10), gridspec_kw={'width_ratios': [3, 5]})
ax[0,0].plot(results_lasso.keys(), results_lasso.values())
ax[0,0].set_xlabel('Lambda')
ax[0,0].set_ylabel('Test Error')
ax[0,0].set_title('Lasso: Lambda vs. Test Error')

df_params_l = pd.DataFrame(reg_path_lasso).T
df_params_l.plot(ax=ax[0,1])
ax[0,1].axvline(min(results_lasso, key=results_lasso.get),color='white')
ax[0,1].set_xlabel('Lambda')
ax[0,1].set_ylabel('coefficient')
ax[0,1].set_title('Regularization paths for Lasso Regression')
ax[0,1].grid()

ax[1,0].plot(results_ridge.keys(), results_ridge.values())
ax[1,0].set_xlabel('Lambda')
ax[1,0].set_ylabel('Test Error')
ax[1,0].set_title('Ridge: Lambda vs. Test Error')

df_params_r = pd.DataFrame(reg_path_ridge).T
df_params_r.plot(ax=ax[1,1])
ax[1,1].axvline(min(results_ridge, key=results_ridge.get),color='white')
ax[1,1].set_xlabel('Lambda')
ax[1,1].set_ylabel('coefficient')
ax[1,1].set_title('Regularization paths for Ridge Regression')
ax[1,1].grid()
```

```python
print(f'The Lasso mean squared error is: {mse_lasso}')
print(f'The R squared for the scikit learn Lasso model is {r2_lasso}')
print(f'The best lambda value for Lasso is {min(results_lasso, key=results_lasso.get)}')
print('\n''The ridge model uses the following beta values for the given predictors:')
ridge_params
```

```python
print(f'The R squared for the scikit learn Ridge model is {r2_ridge}')
print(f'The Ridge mean squared error is: {mse_ridge}')
print(f'The best lambda value for Ridge is {min(results_ridge, key=results_ridge.get)}')
print('\n''The Lasso model is sparse, and give the following beta values for the given predictors (note the zero values)')
lasso_params
```
